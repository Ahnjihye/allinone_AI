**생각 정리하기**
1. 빈 딕셔너리 만들기 (키 = 사용처, 값 = 금액)
2. 카드내역을 한 줄씩 읽으면서 (반복문)
3. 딕셔너리에 사용처가 있으면 금액을 더하고,
4. 딕셔너리에 사용처가 없으면 키-값을 추가한다.
5. 딕셔너리가 완성되면 값(금액)을 기준으로 내림차순 정렬 & 앞에서 10개 데이터만 뽑기
6. 딕셔너리 키(사용처)와 값(사용금액)을 따로 분리
7. 그래프 그리기


````
# 1. 빈 딕셔너리 만들기
dic = {}

for row in data : 
    if row[-1] == '전표매입' : 
        store = row[-4]
        payment = int(row[-3])
        # 3. 딕셔너리 키에 store가 있으면 해당 값에 +금액
        if store in dic.keys() : 
            dic[store] += payment
        # 4. 딕셔너리 키에 store가 없으면 새로운 키-값 추가
        else : 
            dic[store] = payment
        
        print(dic)
````

````
# 5. 딕셔너리가 완성되면 값(금액)을 기준으로 내림차순 정렬 
# 앞에서 10개 데이터만 뽑기
import operator as op

dic = sorted(dic.items(), key = op.itemgetter(1), reverse = True)[:10]
dic
````

````
# 6. 딕셔너리 키(사용처)와 값(사용금액)을 따로 분리
top10_store = []
top10_payment = []

for row in dic : 
    top10_store.append(row[0])
    top10_payment.append(row[1])

print(top10_store)
print(top10_payment)
````

````
# 7. 그래프 그리기
plt.figure(dpi=100)
plt.title('10-12월 지출 TOP 10')
plt.barh(top10_store, top10_payment, color = 'crimson')
plt.show()
````
